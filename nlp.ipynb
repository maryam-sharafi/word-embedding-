```python
import pandas as pd
messages = pd.read_csv('imdb_labelled.txt', sep='\t',
                       names=["messages","label"])
```

```python
messages
```

```python
#data cleaning and preprocessing
import nltk
import re
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('hiking')
```

```python
#apply stopwords and stemming
corpus = []
for i in range(0, len(messages)):
  review = re.sub('[^a-zA-Z]',' ',messages['messages'][i])
  review = review.lower().split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
#creating the bag of words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000, binary=True)
x = cv.fit_transform(corpus).toarray()
```

```python
x.shape
```

```python
y = pd.get_dummies(messages['label'])
y = y.iloc[:,1].values
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
X_train,y_train
```

```python
from sklearn.naive_bayes import MultinomialNB
sentiment_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
#prediction
y_pred = sentiment_detect_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report , accuracy_score
score = accuracy_score(y_test,y_pred)
print(classification_report(y_pred,y_test))
print(score)
```

```python
#creating the TFIDF model
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=2000)
X = tv.fit_transform(corpus).toarray()
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
sentiment_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
#prediction
y_pred = sentiment_detect_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report , accuracy_score
score = accuracy_score(y_test,y_pred)
print(classification_report(y_pred,y_test))
print(score)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_pred,y_test))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
corpus=[]
for i in range(0,len(messages)):
  review = re.sub('[^a-zA-Z]',' ',messages['messages'][i])
  review = review.lower().split()

  review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
##simple_preprocess:convert document into list lowercase
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
nltk.download('punkt')
```

```python
words = []
for sent in corpus:
  sent_token = sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```
